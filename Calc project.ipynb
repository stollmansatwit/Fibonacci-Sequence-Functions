The goal of this project was to find and display three different ways of calculating the fibonacci sequence to the n-th term. The first function is an exponential runtime function meaning the time increases exponentially to calculate the next number in the sequence. The second function is the fastest and it runs in O(log n) time. This function makes use of Python's ability to return 2 integers. The third function creates a list starting at [0,1] and adds the last item plus the second to last item to the list each time. For each function written, I ran it for 50 seconds and plotted the time vs the number in the sequence it was on. By changing the value of my variable called seconds, it is easy to run each function for as short or as long as you want to see how the graph changes over time.

```python
#Very slow algorithm
processes = [] #List to track recurssive calls
def fib(n):
    if n<3: #Base case
        return 1
    else:
        processes.append(1) #Increment count of recurssive calls
        return fib(n-1) + fib(n-2)
```

The length of my list called processes can be printed after calling the fib function to see how many times a recurssive call was made

```python
#Calculating time to run vs number in sequence for slow algorithm
import time
import matplotlib.pyplot as plt
import numpy as np

seconds = 50
sequence = []
nums = []
times = []
startTime = time.time()


for n in range(3000000):
    nums.append(n) #Adds our number to the list nums
    sequence.append(fib(n))
    endTime = time.time()
    totalTime = endTime - startTime
    times.append(totalTime) #Adds our total time to times
    
    
    if totalTime>=seconds:
        break #Stops running once a calculation takes 50 seconds or more
```

```python
len(sequence)
```

We can see that after 50 seconds, my very slow algorithm calculated up to the 42nd number in the sequence.

```python
#Creating npArrays for plotting
nums = np.array(nums)
times = np.array(times)
```

```python
#Plotting
import seaborn as sns
sns.set_theme(style = "whitegrid")
plt.plot(times, nums, linewidth = 2.5, color = 'black')
plt.title("Calculating The n-th number in Fibonacci Sequence over Time", fontsize = 14)
plt.ylabel("Current number in sequence")
plt.xlabel("Time (seconds)")
plt.show()
```

```python
# Fastest algorithm using matrix exponentiation
def fibFaster(n):
    if n == 0:
        return (0,1) # Base case
    else:
        a,b = fibFaster(n//2)
        c = a*(b*2-a)
        d = a*a+b*b
        if n%2 == 0:
            return (c,d)
        else:
            return (d,c+d)
def fibonacci(n):
    return fibFaster(n)[0]
```

```python
#Calculating time to run vs number in sequence for fast algorithm

seconds = 50
sequence = []
nums2 = []
times2 = []
startTime = time.time()
for n in range(30000000):
    
    nums2.append(n) #Adds our number to the list nums
    sequence.append(fibonacci(n))
    
    endTime = time.time()
    totalTime = endTime - startTime
    
    
    times2.append(totalTime) #Adds our total time to the list times
    if totalTime>=seconds:
        break
```

```python
len(sequence)
```

We can see that after 50 seconds, my very fast algorithm calculated up to the 96276th number in the sequence.

```python
#Creating npArrays for plotting
nums2 = np.array(nums2)
times2 = np.array(times2)
```

```python
#Plotting
plt.plot(times2, nums2, linewidth = 2.5, color = 'dodgerblue')
plt.plot(times, nums, linewidth = 2.5, color = 'black')
plt.ylabel("Current number in sequence")
plt.xlabel("Time (seconds)")
plt.title("Calculating The n-th number in Fibonacci Sequence over Time", fontsize = 14)

#Scale
plt.yscale('log') # Using a logarithmic scale so both curves are visible

plt.show()
```

```python
#Fast Algorithm
def fibo(n):
    seq = [0,1]
    while len(seq)<n:
        next_num = seq[-1]+seq[-2]
        seq.append(next_num)
    return seq
```

```python
#Calculating time to run vs number in sequence for fast algorithm

seconds = 50
sequence = []
nums1 = []
times1 = []
startTime = time.time()
for n in range(3000000):
    nums1.append(n) #Adds our number to the list nums
    sequence.append(fibo(n))
    endTime = time.time()
    totalTime = endTime - startTime
    times1.append(totalTime) #Adds our total time to times
    
    
    if totalTime>=seconds:
        break 
```

```python
#Creating npArrays for plotting
nums1 = np.array(nums1)
times1 = np.array(times1)
```

```python
#Plotting
plt.plot(times2, nums2, linewidth = 2.5, color = 'dodgerblue', label = 'fastest')
plt.plot(times1, nums1, linewidth = 2.5, color = 'gold', label = 'fast')
plt.plot(times, nums, linewidth = 2.5, color = 'black', label = 'slow')

plt.ylabel("Current number in sequence")
plt.xlabel("Time (seconds)")

plt.title("Calculating The n-th number in Fibonacci Sequence over Time")

#Scale
plt.yscale('log')
plt.legend()
plt.show()
```
